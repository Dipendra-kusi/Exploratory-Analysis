# Melting Data: changing the column  into rows 

```python

# we want to melt down columns: tortilla_calories, turkey_calories, waffle_calories inoto single column
# here we combine the three column to one as type_calories
pd.melt(frame=df_Food, value_vars=['tortilla_calories', 'turkey_calories', 'waffle_calories'], 
        var_name='type_calories',
        value_name='result_calories')
```

# Pivot: Converting the rows to column

```python
# here we convert the coffee values to column

df_calories =pd.melt(frame=df_Food, value_vars=['tortilla_calories', 'turkey_calories', 'waffle_calories'], 
        var_name='type_calories',
        value_name='result_calories')

df_Food.pivot(columns='coffee',values='GPA').head()
```

```python
# to fix dublicate error in pivot 
# use pivot table instead 

# count the gender, thai food
import numpy as np
df_Food.pivot_table(values='calories_chicken',index='thai_food',columns='Gender',aggfunc='count')
```

```python
# aggregate using mean

df_Food.pivot_table(values='calories_chicken',index='thai_food',columns='Gender',aggfunc=np.mean)
```
